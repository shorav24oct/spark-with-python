```python
resturant_df = spark.read.format("json")\
    .option("inferschema", "true")\
    .option("multiline", "true")\
    .load("/Volumes/workspace/default/sks/resturant_json_data.json")
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
resturant_df.printSchema()
```

```python
resturant_df.select("*", explode("restaurants").alias("new_restaurant"))\
    .drop("restaurants")\
    .select("*", explode("new_restaurant.restaurant.location").alias("new_location"))\
    .select("new_restaurant.R.res_id", "new_restaurant.name", "new_location.city")\
    .drop("new_restaurant")\
    .show()
```
